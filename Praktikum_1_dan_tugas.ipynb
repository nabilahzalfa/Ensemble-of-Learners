## Import Libraray

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # import DT
from sklearn.ensemble import RandomForestClassifier # import RandomForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

## Persiapan Data

```python
df = pd.read_csv('data/wbc.csv')
df.head()
```

```python
# Cek kolom null
df.isnull().sum()
```

```python
# Seleksi fitur

# Slice dataframe mulai dari kolom 'radius_mean' sampai 'fractal_dimension_worst'
X = df.iloc[:,2:-1]
y = df['diagnosis']
y = y.map({'M':1, 'B':0}) # Encode label

# Cek jumlah fitur dan instance
X.shape
```

## Split data training dan testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Traning Decision Tree

```python
# Secara default, DecisionTreeClassifier dari scikit-learn akan menggunakan nilai "Gini" untuk kriteria
# Terdapat beberapa "hyperparamater" yang dapat digunakan. Silahka baca dokumentasi
# Pada kasus ini kita akan menggunakan parameter default
dt = DecisionTreeClassifier()

# Sesuaikan dt ke set training
dt.fit(X_train, y_train)

# Memprediksi label set test
y_pred_dt = dt.predict(X_test)

#  menghitung set accuracy
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")
```

## Training RandomForest

```python
# Pada kasus kali ini kita akan menggunakan estimator pada RandomForest
# Untuk detail parameter (hyperparameter) silahkan cek dokumentasi

rf = RandomForestClassifier(n_estimators=10, random_state=1)

# Sesuaikan dt ke set training
rf.fit(X_train, y_train)

# Memprediksi label set test
y_pred_rf = rf.predict(X_test)

#  menghitung set accuracy
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Test set accuracy: {:.2f}".format(acc_rf))
print(f"Test set accuracy: {acc_rf}")
```

# TUGAS PRAKTIKUM 1

## Persiapan Data

```python
df = pd.read_csv('data/mushrooms.csv')
df = df.reset_index()
df.head()
```

```python
# Data untuk label 
new_labels = {
    'a':0.1,'b':0.2,'c':0.3,'d':0.4,'e':0.5,'f':0.6,'g':0.7,'h':0.8,'i':0.9,'j':1.0,'k':1.1,'l':1.2,'m':1.3,
    'n':1.4,'o':1.5,'p':1.6,'q':1.7,'r':1.8,'s':1.9,'t':2.0,'u':2.1,'v':2.2,'w':2.3,'x':2.4,'y':2.5,'z':2.6
    }

# Encode label
df['cap-shape'] = df['cap-shape'].map(new_labels)
df['cap-surface'] = df['cap-surface'].map(new_labels)
df['cap-color'] = df['cap-color'].map(new_labels)
df['bruises'] = df['bruises'].map(new_labels)
df['odor'] = df['odor'].map(new_labels)
df['gill-attachment'] = df['gill-attachment'].map(new_labels)
df['gill-spacing'] = df['gill-spacing'].map(new_labels)
df['gill-size'] = df['gill-size'].map(new_labels)
df['gill-color'] = df['gill-color'].map(new_labels)
df['stalk-shape'] = df['stalk-shape'].map(new_labels)
df['stalk-root'] = df['stalk-root'].map(new_labels)
df['stalk-surface-above-ring'] = df['stalk-surface-above-ring'].map(new_labels)
df['stalk-surface-below-ring'] = df['stalk-surface-below-ring'].map(new_labels)
df['stalk-color-above-ring'] = df['stalk-color-above-ring'].map(new_labels)
df['stalk-color-below-ring'] = df['stalk-color-below-ring'].map(new_labels)
df['veil-type'] = df['veil-type'].map(new_labels)
df['veil-color'] = df['veil-color'].map(new_labels)
df['ring-number'] = df['ring-number'].map(new_labels)
df['ring-type'] = df['ring-type'].map(new_labels)
df['spore-print-color'] = df['spore-print-color'].map(new_labels)
df['population'] = df['population'].map(new_labels)
df['habitat'] = df['habitat'].map(new_labels)

# Cek data
df.head()
```

```python
# Cek kolom null
df.isnull().sum()
```

```python
# Seleksi fitur

# Slice dataframe mulai dari kolom 'cap-shape' sampai 'habitat'
X = df.iloc[:,2:-1]
y = df['class']
y = y.map({'p':1, 'e':0}) # Encode label

# Cek jumlah fitur dan instance
X.shape
```

## Split data training dan testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Menggantikan nilai tak terhingga dengan nilai hingga yang tinggi dan nilai nan dengan angka
X_train = np.nan_to_num(X_train)
y_train = np.nan_to_num(y_train)
X_test = np.nan_to_num(X_test)
y_test = np.nan_to_num(y_test)
```

## Traning Decision Tree

```python
# Secara default, DecisionTreeClassifier dari scikit-learn akan menggunakan nilai "Gini" untuk kriteria
# Terdapat beberapa "hyperparamater" yang dapat digunakan. Silahka baca dokumentasi
# Pada kasus ini kita akan menggunakan parameter default
dt = DecisionTreeClassifier()

# Sesuaikan dt ke set training
dt.fit((X_train), y_train)

# Memprediksi label set test
y_pred_dt = dt.predict(X_test)

#  menghitung set accuracy
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")
```

## Training RandomForest

```python
# Pada kasus kali ini kita akan menggunakan estimator pada RandomForest
# Untuk detail parameter (hyperparameter) silahkan cek dokumentasi

rf = RandomForestClassifier(n_estimators=10, random_state=1)

# Sesuaikan dt ke set training
rf.fit(X_train, y_train)

# Memprediksi label set test
y_pred_rf = rf.predict(X_test)

#  menghitung set accuracy
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Test set accuracy: {:.2f}".format(acc_rf))
print(f"Test set accuracy: {acc_rf}")
```
